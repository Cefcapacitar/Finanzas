```python
questions = [
    ("1. What is your age?", "Under 30 = 5, 30–45 = 4, 46–60 = 3, Over 60 = 1"),
    ("2. What is your investment horizon?", "Less than 3 years = 1, 3–7 years = 3, More than 7 years = 5"),
    ("3. What is your main investment goal?", "Preserve capital = 1, Balance growth and safety = 3, Maximize growth = 5"),
    ("4. How stable is your income?", "Very stable = 5, Somewhat stable = 3, Unstable = 1"),
    ("5. Do you have an emergency fund?", "Yes (6+ months) = 5, Partial (3–6 months) = 3, No = 1"),
    ("6. What percentage of your income do you save monthly?", "Above 20% = 5, 10–20% = 3, Below 10% = 1"),
    ("7. How would you describe your debt level?", "Low/none = 5, Moderate = 3, High = 1"),
    ("8. What is your net worth compared to your liabilities?", "High = 5, Medium = 3, Low = 1"),
    ("9. How do you react to a 10% portfolio loss?", "Stay calm = 5, Slightly worried = 3, Very anxious = 1"),
    ("10. How do you react to a 20% loss?", "Hold or buy more = 5, Wait and see = 3, Sell = 1"),
    ("11. How experienced are you with investments?", "Very experienced = 5, Some experience = 3, Beginner = 1"),
    ("12. How often do you check your portfolio?", "Rarely = 5, Sometimes = 3, Daily = 1"),
    ("13. How do you decide on investments?", "Based on research = 5, Based on advice = 3, Based on emotions = 1"),
    ("14. Do you follow a financial plan?", "Yes, strictly = 5, Sometimes = 3, No = 1"),
    ("15. How do you handle market volatility?", "Comfortable = 5, Neutral = 3, Uncomfortable = 1"),
    ("16. How diversified is your current portfolio?", "Highly diversified = 5, Some diversification = 3, Little or none = 1"),
    ("17. What return do you expect annually?", "Over 10% = 5, 5–10% = 3, Under 5% = 1"),
    ("18. How much of your portfolio could you lose without changing strategy?", "Over 20% = 5, 10–20% = 3, Under 10% = 1"),
    ("19. How do you prefer to invest?", "Stocks/ETFs = 5, Mixed funds = 3, Bonds/Cash = 1"),
    ("20. How much financial news do you follow?", "Regularly = 5, Occasionally = 3, Rarely = 1")
]

# Collect user input
scores = []
print("Please answer each question by entering the corresponding score (1, 3, or 5).\n")

for q, opts in questions:
    print(f"{q}\nOptions: {opts}")
    while True:
        try:
            score = int(input("Your score (1, 3, or 5): "))
            if score in [1, 3, 5]:
                scores.append(score)
                print()
                break
            else:
                print("Please enter a valid score: 1, 3, or 5.")
        except ValueError:
            print("Invalid input. Please enter a number (1, 3, or 5).")

# Calculate total and classify
total_score = sum(scores)
print(f"\nYour total score is: {total_score}")

if total_score <= 30:
    profile = "Conservative"
elif total_score <= 60:
    profile = "Moderate"
else:
    profile = "Aggressive"

print(f"Your Risk Profile: {profile}")


```
